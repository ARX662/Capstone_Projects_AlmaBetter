### Variables Description

# Description of the Variable which would be usefull during our analysis.
**Event ID (eventid):** A unique identifier for each terrorist event.

**Month (imonth):** The month of the event.

**Region Name (region_txt):** The name of the region.

**Location (location):** Specific location of the event (e.g., city, province).

**Summary (summary):** A brief description of the event.

**Attack Type (attacktype1):** Numeric code representing the type of attack (e.g., bombing, assassination).

**Attack Type Name (attacktype1_txt):** The name of the attack type.

**Target Type (targtype1):** Numeric code representing the target type (e.g., civilians, government).

**Target Type Name (targtype1_txt):** The name of the target type.
Perpetrator Group (gname): The name of the terrorist group responsible for the attack.

**Number of Perpetrators (nperps):** The count of perpetrators involved.

**Number of Perpetrators Captured (nperpcap):** The count of captured perpetrators.

**Number of Victims Killed (nkill):** The number of people killed in the attack.

**Number of Victims Wounded (nwound):** The number of people wounded in the attack.

**Property Damage (property):** Indicates whether property was damaged.

**Hostage Situation (ishostkid):** Indicates whether hostages were taken.

**Number of Fatalities (nkill):** Total fatalities in the incident.

**Number of Injured (nwound):** Total people injured in the incident.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# renaming columns to make it easier for analysis
df.rename(columns={'iyear':'year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type'},inplace=True)
````

````python
# filling missing values in wounded and no. of killed columns and having data type as int to make analysis accurate.
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df["killed"] = df["killed"].fillna(0).astype(int)
# creating a column casualties to better understand the damage casued
df["Casulaties"] = df['Wounded'] + df["killed"]
````

````python
#checking if fields required for analysis have missing values changes made
print("Country",df['Country'].isnull().sum())
print("Region",df['Region'].isnull().sum())
print("Attack_type", df['AttackType'].isnull().sum())
print("Killed",df['killed'].isnull().sum())
print("wounded", df['Wounded'].isnull().sum())
print("Group",df['Group'].isnull().sum())
print("hostages", df["ishostkid"].isnull().sum())
print("property",df["property"].isnull().sum())
````

Hostages values are ignored as the value 0 represents no hostages taken which is being count as missing values by the isnull() method

### What all manipulations have you done and insights you found?

Wounded and Killed had NAN values which were replaced by 0 as the fields are not numeric nature.Futhermore the names of the fields were changed making it easier to understand and use

#### Chart - 2

````python
# Chart - 2 Attacks over the year by region by creating cross table with regionwise attacks over the years
cross_tab_reg = pd.crosstab(df["year"], df["Region"])
cross_tab_reg = cross_tab_reg[cross_tab_reg.sum().sort_values(ascending =False).index]
# Define colorblind-friendly colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
cross_tab_reg.plot(kind = "line",figsize = (13,6), color=colors)
plt.title("Number of attacks over the years by region")
plt.ylabel("Number of attacks")
plt.show()
````

##### 1. Why did you pick the specific chart?

###*Line charts allow us to observe trends and patterns over time.By connecting data points with lines, we can easily identify increases, decreases, or fluctuations in the number of attacks for each region.Each region’s line can be visually compared to others on the same chart.We can pinpoint specific years with high or low attack rates.*

##### 2. What is/are the insight(s) found from the chart?

###**General Increase:** Overall, there is a general increase in the number of attacks across all regions over time.

###**Sharp Rise in Middle East & North Africa and South Asia:** Since around 2000, the Middle East & North Africa and South Asia regions have seen a significant surge in attacks.

###**Western Europe Decline:** Western Europe had a peak in attacks around the mid-1980s but has since experienced a substantial decline.

### **Central America & Caribbean:** This region had a high number of attacks during the 1980s but decreased afterward.

### **Other Regions:** Sub-Saharan Africa and Southeast Asia show an increase, but not as pronounced as in the Middle East & North Africa or South Asia.
